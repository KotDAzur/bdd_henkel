```python
# 1. Évolution des ventes par famille de produits
evolution_ventes_famille = (sell_in_bricomarche
                            .groupby('famille')[['total_annuel_2022', 'total_annuel_2023']]
                            .sum()
                            .reset_index())
evolution_ventes_famille['evolution(%)'] = ((evolution_ventes_famille['total_annuel_2023'] - evolution_ventes_famille['total_annuel_2022']) / evolution_ventes_famille['total_annuel_2022']) * 100

# 2. Top 5 des familles de produits avec la plus forte croissance et décroissance
top_5_growth_famille = evolution_ventes_famille.sort_values(by='evolution(%)', ascending=False).head(5)
top_5_decline_famille = evolution_ventes_famille.sort_values(by='evolution(%)', ascending=True).head(5)

# 3. Répartition des visites annuelles par ville
visites_par_ville = (visite_annuelle_bricomarche
                     .groupby('Ville')[['Visites2022', 'Visites2023']]
                     .sum()
                     .reset_index())

# 4. Budget total accordé par type d'accord
budget_par_type_accord = (pp_22_23_bricomarche
                          .groupby('type_accord')['budget_accord_ht']
                          .sum()
                          .reset_index())

# Affichage des résultats
print("Évolution des ventes par famille de produits:")
print(evolution_ventes_famille)
print("\nTop 5 des familles de produits avec la plus forte croissance:")
print(top_5_growth_famille)
print("\nTop 5 des familles de produits avec la plus forte décroissance:")
print(top_5_decline_famille)
print("\nRépartition des visites annuelles par ville:")
print(visites_par_ville)
print("\nBudget total accordé par type d'accord:")
print(budget_par_type_accord)
```
